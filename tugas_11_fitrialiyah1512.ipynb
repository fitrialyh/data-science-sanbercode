```python
#persiapkan semua library pada sel ini
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
import string

import nltk
nltk.download('stopwords')
nltk.download('punkt')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
#buka berkas tugas dan masukkan dalam suatu variabel/dataframe
#tampilkan datanya
df = pd.read_csv('text_cleaning.csv')
df
```

```python
#buat sebuah fungsi case folding yang memiliki fungsi di bawah
'''
1. lowercase
2. menghilangkan angka
3. menghilangkan tanda baca
4. menghilangkan tautan
5. menghilangkan whitespaces
'''

def casefolding(text):
  # 1. lowercase
  text = text.lower()

  # 2. menghilangkan angka
  text = re.sub(r"\d+", "", text)

  # 3. menghilangkan tanda baca
  text = text.translate(str.maketrans("","",string.punctuation))
  
  # 4. menghilangkan tautan
  text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
  text =  text.replace("http://", " ").replace("https://", " ")

  # 5. menghilangkan whitespaces
  text = text.strip()

  return text
```

```python
#aplikasikan fungsi pada kolom artikel dan taruh hasilnya pada kolom clean
#tampilkan

df['clean'] = df['artikel'].apply(casefolding)
df
```

```python
text = "Nick likes to play football, however he is not too fond of tennis."
text_tokens = word_tokenize(text)
print(text_tokens)
```

```python
#buat sebuah fungsi untuk menghilangkan stopwords
'''
menghilangkan stopword bahasa indonesia + kata bandung dan kota
'''
stop_words = stopwords.words('indonesian')
stop_words.extend(['bandung', 'kota'])
stop_words = set(stop_words)

word_tokens = df['clean'].apply(word_tokenize)
df['clean'] = [' '.join(map(str, l)) for l in word_tokens.apply(lambda x: [item for item in x if item not in stop_words])]
df
```

```python
#buat sebuah fungsi untuk melakukan stemming
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(text):
  text = stemmer.stem(text)
  return text
```

```python
#aplikasikan fungsi pada kolom clean dan taruh hasilnya pada kolom clean
#tampilkan

df['clean'] = df['clean'].apply(stemming)
df
```
